### 1. Which establishments have a hygiene score equal to 20?

```python
# Create a query that finds establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}
results = establishments.find(query)

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query) 
print(f"Number of establishments with hygiene score 20: {count}")

# Display the first document in the results using pprint
first_document = results[0] if count > 0 else None
pprint(first_document)
```

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': 20} 

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query) 
print(f"Establishments had {count} Hygiene with score 20\n")

# Display the first document in the results using pprint
pprint(establishments.find_one(query))
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
london_query = {
                'LocalAuthorityName': {'$regex': 'London'},
                'RatingValue': {'$gte': 4}
                }

# Use count_documents to display the number of documents in the result
london_query_count = establishments.count_documents(london_query)
print(f"Number of establishments in London with RatingValue >= 4: {count}")

# Display the first document in the results using pprint
pprint(establishments.find_one(london_query))
```

```python
# Convert the result to a Pandas DataFrame
london_query_df = pd.DataFrame(establishments.find(london_query))

# Display the number of rows in the DataFrame
print("Number of rows in the DataFrame: ", len(london_query_df))

# Display the first 10 rows of the DataFrame
london_query_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.49042
longitude = 0.08384

nearest_query = {'geocode.latitude': {'$lte':(latitude + degree_search)},
                 'geocode.latitude': {'$gte':(latitude - degree_search)},
                 'geocode.longitude': {'$lte': (longitude + degree_search)},
                 'geocode.longitude': {'$gte':(longitude - degree_search)},
                 'RatingValue': 5
                }

sort = [('establishments.scores.Hygiene', 1)]

# Print the results
for result in establishments.find(nearest_query).sort(sort).limit(5):
    pprint(result)
```

```python
# Convert result to Pandas DataFrame
nearest_query_df = pd.DataFrame(establishments.find(nearest_query))
nearest_query_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
local_authority_query = {'$match': {'scores.Hygiene': 0}}

# 2. Groups the matches by Local Authority
local_authority_groups_query = {'$group': {'_id': '$LocalAuthorityName', 'count': {'$sum': 1}}}
 
# 3. Sorts the matches from highest to lowest
local_authority_sort = {'$sort': { 'count': -1}}

# Combine the stages into a pipeline
pipeline = [
    local_authority_query,
    local_authority_groups_query,
    local_authority_sort
]

# Run the aggregation query
local_authority_results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print(f"establishments had {len(local_authority_results)} with a hygiene score of 0")

# Print the first 10 results
pprint(local_authority_results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
local_authority_result_df = pd.DataFrame(local_authority_results)

# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(local_authority_result_df))

# Display the first 10 rows of the DataFrame
local_authority_result_df.head(10)
```
